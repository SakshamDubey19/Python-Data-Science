# Graphs

1. Univariate Graphs : Graphs that take a single variable
    - Line plot - for growth/decay over a period
    - Histogram - for distribution of data 
    - Density plot - for distribution of data in normalised form
    - box plot  - for distribution of data in qyartiles
    - Pie chart - for discrete data comparison

2. Bivariate Graph : Graphs that take two variables
    - Scatter plot - relation between two variables
    - Heat map - relation between two variables in matrix form
    - Violin plot - realtion between two variables in density form
    - Pair plot - relation between 2 variables in grid form

3. Multivariate Graph - Graphs that take more than two
    - Bubble plot - realtion between 2 variables with size
    - 3D Scatter plot - relation betweeen 3 variables on x,y,z axis
    - Stacked bar chart - realtion between 3 variables in bar form

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

# Histogram

```python
x = np.random.normal(size = 1000)
plt.hist(x, bins=np.arange(-3,3,.5), edgecolor = 'black')
plt.show


sns.histplot(x, bins=np.arange(-3,3,.5), kde=True)
```

```python
px.histogram(x, nbins=20)
```

# BOX PLOT

```python
x = np.random.normal(size=100)
plt.boxplot(x)
plt.show
```

```python
sns.boxplot(x)
```

```python
px.box(x)
```

# Pie Chart

```python
rainfall = [24,45,35,23,65,23]
labels = ['jan','feb','march','april','may','june']
plt.pie(
    rainfall,
    autopct='%.2f%%',
    labels=labels
)
plt.show
```

```python
px.pie(
    values=rainfall,
    names=labels,
    hole=.25     #for donut chart
)
```

```python
# Area chart
x= np.arange(20)
y = np.random.randint(2,20,20)
plt.fill_between(x,y) #area chart
plt.show
```

```python
px.area(x=x,y=y)
```

# Bar chart

```python
years  = np.arange(2010,2020, dtype=int)
sales  = np.random.randint(1000,5000,10)
plt.bar(years,sales)
```

```python
sns.barplot(x= years, y=sales)
```

```python
px.bar(x=years,y=sales)
```

```python
px.bar_polar(r=years,theta=sales)
```

# SCATTER PLOT

```python
x = np.random.normal(size=1000)
y= np.random.normal(size=1000)
plt.scatter(x,y, alpha=.5, edgecolors='black')
```

```python
sns.scatterplot(x=x,y=y)
```

```python
px.scatter(x=x,y=y)
```

# Heat map

```python
sns.heatmap(np.random.rand(5,5), annot=True)
```

```python
px.density_heatmap(x=x,y=y)
```

```python
plt.contour(np.random.rand(10,10))
```

# Violin Plot

```python
x= np.random.normal(size=100)
plt.violinplot(x)
plt.show
```
